you can use shift on a data frame or timeseries.

shifting can be used in financial analysis where you'd like to see changes in sales by day or across a time span

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('FB.csv')
```

```python
df.head()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df = df.iloc[:, 0:4]
```

```python
df.head()
```

```python
df['Prev_Close'] = df['Close'].shift(1)
```

```python
df.head()
```

```python
df.drop(columns=['High', 'Low'], inplace=True)
```

```python
df.head()
```

```python
df['Close_Diff'] = df['Close'] - df['Prev_Close']
```
